```python
# Reinforcement learning (RL)
# What is reinforcement learning?
# Reinforcement learning is a machine learning approach to learning from experience.
# Key conepts of reinforcement learning

# 1. Agent: learner or decision maker
# 2. Environment: context in which the agent operates
# 3. Reward: Feedback from the environment based on the action taken by the agent
# 4. Action:A decision made by the agent to interact with the environment
# 5. State: The current state of the environment
# 6.Policy: A function that maps the state to the action
# 7. value function: A function that maps the state to the value of the state

# Key characteristics of reinforcement learning
# No labelled data provided
# They focus on learning from experience
# involves exploration and exploitation
# The model identify hidden patterns , structures or relation in the data
# Used for EDA exploratory data analysis
# Can handle large volumes of data

# Reinforcement learning algorithms
# Q-learning
# SARSA
# Expected SARSA

# Q- learning 
# Environment (Position, Goal, Reward)
# Reward(Positive for reaching the goal and negative for not reaching the goal)

# Create a simple Q-learning agent that learns to navigate a grid environment to reach a goal while avoiding walls

import numpy as np
import random

# Define the environment
position = 5 #Initial position of the agent 0 to 4
actions = 2 # 0: move left, 1: move right 

# initialize Q-table
Q = np.zeros((position, actions))

# Define the parameters 
episodes = 10000 # Number of episodes to train the agent
learning_rate = 0.8 # Learning rate
gama = 0.9 # Discount factor
epislon = 0.3 # Probability of taking a random action


# Training the loop 
for episode in range(episodes):
    state = random.randint(0, position - 1) # Random initial state
    
# Action selection(Epsilon-greedy policy)
    if random.randint(0, 1) < epislon:
        action = random.randint(0, actions - 1)
    else: # move right
        action = np.argmax(Q[state]) # exploit the best known action

#  Take action
    if action == 0: # move left
        new_state = max(0, state - 1)
    else: # move right
        new_state = min(position -1, state + 1)
        
# Reward structure
    if new_state == position - 1:
        reward = 10
    else:
        reward = -1
        
# Q-learning update rule
    Q[state][action] = Q[state][action] + learning_rate * (reward + gama * np.max(Q[new_state]) - Q[state][action])
    
    
# Transition to the next state   
    state = new_state # update the state to the next state

# show learned Q-table
print('Q-table:')
print(Q)

# Test the train agent
state = 0 # start from the initial position
steps = 0 # initialize the step counter
print('\n Agent path goal')

# Agent path to the goal
while state < position - 1:
    steps += 1
    action = np.argmax(Q[state]) # exploit the best known action
    if action == 0:
        new_state = max(0, state - 1)
    else:
        new_state = min(position -1, state + 1)
    print('step', steps, 'position', state, 'action', action)
    
# Final Q-table
print('\n Final Q-table:')
print(Q)

```
